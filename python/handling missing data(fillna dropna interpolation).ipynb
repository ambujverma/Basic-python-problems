```python
import pandas as pd
df = pd.read_csv("dataframe.csv",parse_dates=["day"])
df.set_index('day' ,inplace=True)
```

```python
df
```

```python
# method for fill the NaN values
new_df=df.fillna(0)
new_df
```

```python
#using dict for event section cos 0 is meaningless
new_df=df.fillna({
    'temprature':0,
    'windspeed':0,
    'event':'no event'
})
new_df
```

```python
#new_df=df.ffill(). also using bfill method
new_df =df.fillna(method='ffill', axis="rows")
new_df
```

```python
new_df=df.interpolate()
new_df
```

```python
#but there is not right interpolate for all days c/s 1/1 ot 1/4 direct so we using time wise polation
new_df=df.interpolate(method='time')
new_df
```

```python
# thresh is using for remove the rows with number of NaN values
new_df=df.dropna(thresh=3)
new_df
```

```python
# how to insert missing date using date range
dt=pd.date_range("01-01-2019" ,"01-11-2019")
idx =pd.DatetimeIndex(dt)
df=df.reindex(idx)
df
```
